```python
# grouping results
df_gptest = df[['drive-wheels','body-style','price']]
grouped_test1 = df_gptest.groupby(['drive-wheels','body-style'],as_index=False).mean()
grouped_test1
```

### Pivot Table

```python
pivot_table = grouped_test1.pivot(index='drive-wheels',columns='body-style')
pivot_table
```

<p>Often, we won't have data for some of the pivot cells. We can fill these missing cells with the value 0.


```python
pivot_table = pivot_table.fillna(0)
pivot_table
```

```python
df.groupby(['body-style'])['price'].mean()
```

<h4>Variables: Drive Wheels and Body Style vs. Price</h4>


Let's use a heat map to visualize the relationship between Body Style vs Price.


```python
#use the grouped results
plt.pcolor(pivot_table, cmap='RdBu')
plt.colorbar()
plt.show()
```

<p>The heatmap plots the target variable (price) proportional to colour with respect to the variables 'drive-wheel' and 'body-style' on the vertical and horizontal axis, respectively. This allows us to visualize how the price is related to 'drive-wheel' and 'body-style'.</p>

<p>The default labels convey no useful information to us. Let's change that:</p>


```python
fig, ax = plt.subplots()
im = ax.pcolor(pivot_table, cmap='RdBu')

#label names
row_labels = pivot_table.columns.levels[1]
col_labels = pivot_table.index

#move ticks and labels to the center
ax.set_xticks(np.arange(pivot_table.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(pivot_table.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.show()
```

## Correlation and Causation


<p>Pearson Correlation is the default method of the function "corr".


```python
df[['wheel-base','price']].corr()
```

```python
from scipy import stats
```

<h3>Wheel-Base vs. Price</h3>


Let's calculate the  Pearson Correlation Coefficient and P-value of 'wheel-base' and 'price'.


```python
pearson_coef = stats.pearsonr(df['wheel-base'], df['price'])
pearson_coef.pvalue
```

<h4>Conclusion:</h4>
<p>Since the p-value is $<$ 0.001, the correlation between wheel-base and price is statistically significant, although the linear relationship isn't extremely strong (~0.585).</p>


<h3>Horsepower vs. Price</h3>
